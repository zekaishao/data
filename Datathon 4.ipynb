```python
import numpy as np 
import pandas as pd 
```

```python
data = pd.read_csv("training_v2.csv")
```

```python
df = data[['hospital_death', 'age', 'bmi', 'gender', 'ethnicity', 'glucose_apache', 'heart_rate_apache', 'diabetes_mellitus', 'apache_2_bodysystem']]
df.head()
```

```python
len(df)
```

```python
df = df.dropna()
```

```python
len(df)
```

```python
df_1 = df.copy()
df_1
```

```python
bins_age = [10, 20, 30, 40, 50, 60, 70, 80, 90]
bins_age
# a_bins = df.a.groupby(cut(df.a,bins))
```

```python
#Adding labels in 
df_1['age_group'] = pd.cut(df_1.age, bins_age)
df_1.head()
```

```python
bins_bmi = [1, 18.5, 25.0, 30.0, 40.0, 99.9]
bins_bmi
```

```python
#generating the labels for bins
bmi_labels = ('Underweight','Normal', 'Overweight', 'Obesity', 'Extreme Obesity')
```

```python
df_1['bmi_group'] = pd.cut(df_1.bmi, bins_bmi, right=False, labels =bmi_labels)
df_1.head()
```

```python
bins_glucose = [1, 60, 80, 140, 200, 350,999]
bins_glucose
```

```python
#generating the labels for bins
glucose_labels = ('Hypoglycemia','Early hypoglycemia', 'Normal', 'Pre Diabetes', 'Diabetes', 'Extreme Diabetes')
```

```python
df_1['glucose_group'] = pd.cut(df_1.glucose_apache, bins_glucose, right=False, labels =glucose_labels)
df_1.head()
```

```python
df_age = df_1[['apache_2_bodysystem', 'age_group']]
df_age = df_age.replace('Undefined diagnoses', 'Undefined Diagnoses')
```

```python
# df_age.index = df_age['apache_2_bodysystem']
```

```python
df_age
```

```python
df_age = pd.crosstab(df_age['apache_2_bodysystem'], df_age['age_group'], rownames=['diagnosis '], colnames=['age group'])
df_age
```

```python
df_bmi = df_1[['age_group', 'bmi_group']]
# df_bmi.index = df_bmi['age_group']
df_bmi = pd.crosstab(df_bmi['age_group'], df_bmi['bmi_group'], rownames=['age group'], colnames=['bmi group'])
df_bmi
```

```python
df_glucose = df_1[['bmi_group', 'glucose_group']]
# df_glucose.index = df_glucose['bmi_group']
df_glucose = pd.crosstab(df_glucose['bmi_group'], df_glucose['glucose_group'], rownames=['bmi group'], colnames=['glucose_group'])
df_glucose
```

```python
df_mortality = df_1[['glucose_group', 'hospital_death']]
# df_mortality.index = df_mortality['glucose_group']

df_mortality = pd.crosstab(df_mortality['glucose_group'], df_mortality['hospital_death'], rownames=['glucose_group'], colnames=['mortality'])
df_mortality
```

```python
display(df_age.stack().values)
display(df_bmi.stack().values)
display(df_glucose.stack().values)
display(df_mortality.stack().values)
```

```python
values = np.concatenate((df_age.stack().values, df_bmi.stack().values, df_glucose.stack().values, df_mortality.stack().values))
values
```
